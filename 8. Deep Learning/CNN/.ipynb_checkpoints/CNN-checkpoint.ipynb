```python
import tensorflow as tf
from keras_preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Preprocessing the Training set
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory(
        'dataset/training_set',
        target_size=(60, 60),
        batch_size=32,
        class_mode='binary')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
        'dataset/test_set',
        target_size=(60, 60),
        batch_size=32,
        class_mode='binary')
```

```python
# Part 2 - Building the CNN

# Initialising the CNN
cnn=tf.keras.models.Sequential()
```

```python
# Step 1 - Convolution
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=4,activation='relu',input_shape=[60,60,3]))
```

```python
# Step 2 - Pooling
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=4,activation='relu',input_shape=[60,60,3]))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=4,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
#step 3 flattening
cnn.add(tf.keras.layers.Flatten())
```

```python
# Step 4 - Full Connection
cnn.add(tf.keras.layers.Dense(units=256,activation='relu'))
```

```python
#step 5 output
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
# Compiling the CNN
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Training the CNN on the Training set and evaluating it on the Test set
cnn.fit(x = training_set, validation_data = test_set, epochs = 50)
```

```python

from keras_preprocessing import image
test_image=image.load_img("dataset/single_prediction/cat_or_dog_1.jpg",target_size=(60,60))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
prediction=cnn.predict(test_image)
training_set.class_indices
if prediction[0][0] == 1:
    result = 'dog'
else:
    result = 'cat'
print(result)
```
