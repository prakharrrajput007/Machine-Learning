```python
###importing libraries
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#importing dataset
```

```python
dataset=pd.read_csv('Salary_Data.csv')
```

```python
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
```

```python
#splitting dataset into training and test
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_Y,test_Y=train_test_split(X,Y,test_size=.20,random_state=0)
```

```python
##now training our linear regression model with the help of training set 
```

```python
from sklearn.linear_model import LinearRegression  ##it is use for building linear regression model
lin_regressor = LinearRegression()
lin_regressor.fit(train_X,train_Y)
```

```python
##predicting the test result using our  model
```

```python
pred_Y=lin_regressor.predict(test_X)
```

```python
## visualising the training set result 
```

```python
plt.scatter(train_X,train_Y,color = 'green')
plt.plot(train_X, lin_regressor.predict(train_X),color='blue')
plt.title('Sal/Exp graph(training set)')
plt.xlabel('Experience in Years')
plt.ylabel('salary of the eployeee')
plt.show()
```

```python
#visualising the test set results
```

```python
plt.scatter(test_X,test_Y,color = 'red')
plt.plot(train_X, lin_regressor.predict(train_X),color='blue')
plt.title('Sal/Exp graph(training set)')
plt.xlabel('Experience in Years')
plt.ylabel('salary of the eployeee')
plt.show()
```
