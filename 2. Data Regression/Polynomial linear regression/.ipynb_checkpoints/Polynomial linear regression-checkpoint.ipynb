## Polynomail Linear Regression
### y=const+a1.x^1+a2.x^2+a3.x^3+..........+anx^n

#### we cant just use linear here because the slope is increasing or der=creasing continuously so it wouldnt be great to represent in linear so we use polynomial regression here 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt ##importing libraries done 
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,-1].values   ##importing dataset done
```

```python
from sklearn.linear_model import LinearRegression
lin_regression = LinearRegression()
lin_regression.fit(X,Y)    #training our linear model done
```

```python
####now we're gonna build polynomial regression model 
from sklearn.preprocessing import PolynomialFeatures
Pol_regression = PolynomialFeatures(degree=4)# the more the degree the more the power variables and the more the curve is touchibg your points but it becomes less smooth
Poly_X = Pol_regression.fit_transform(X) # we using fit_transform because we want to transform our matrix of 1 feature into matrix of 2 features one is x and the second is x^2
lin_regression_2 = LinearRegression()
lin_regression_2.fit(Poly_X,Y)
```

```python
## now lets vsiualise our Linear Regression results
plt.scatter(X,Y,color='green')
plt.plot(X,lin_regression.predict(X),color='blue')
plt.title("BOSS (linear)")
plt.xlabel("POSITIONS LEVEL")
plt.ylabel("Salary")
plt.show()
```

```python
## now visualising the polynomial regression 
plt.scatter(X,Y,color='red')
plt.plot(X,lin_regression_2.predict(Pol_regression.fit_transform(X)),color='blue')
plt.title("BOSS (polynomial)")
plt.xlabel("POSITIONS LEVEL")
plt.ylabel("Salary")
plt.show()
```

```python
#visualising the polynomial with soother curve this time and high resolution 
X_grid = np.arange(min(X), max(X), 0.1) ## instaed of taking level with .5 we're going to consider .1 .2 .3 ...nd so on
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, Y, color = 'red')
plt.plot(X_grid, lin_regression_2.predict(Pol_regression.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# lets predict our from our models 
# for linear regression
print(lin_regression.predict([[7.5]])) ##double square represents 2d array first dimension is for rows and second dimension is for column
print(lin_regression.predict([[6.5]]))
```

```python
print(lin_regression_2.predict(Pol_regression.fit_transform([[7.5]])))
print(lin_regression_2.predict(Pol_regression.fit_transform([[6.5]])))
```
